```python
X_train = pickle.load(open('/content/gdrive/MyDrive/NLP Final Project/model_data_melissa/X_train_res.pkl', 'rb'))
y_train = pickle.load(open('/content/gdrive/MyDrive/NLP Final Project/model_data_melissa/y_train_res.pkl', 'rb'))
X_test = pickle.load(open('/content/gdrive/MyDrive/NLP Final Project/model_data_melissa/X_test.pkl', 'rb'))
y_test = pickle.load(open('/content/gdrive/MyDrive/NLP Final Project/model_data_melissa/y_test.pkl', 'rb'))
```

```python
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import tensorflow as tf 

y_train_labels = pd.get_dummies(y_train).values
y_test_labels = pd.get_dummies(y_test).values

def create_model(hiddennodes=1):
    # create model
    seed = 123
    np.random.seed(seed)
    tf.random.set_seed(seed)

    model = Sequential()
    model.add(Dense(hiddennodes, input_dim = 307, activation = 'relu')) # tuning in first hidden layer
    model.add(Dense(100, activation = 'relu')) 
    model.add(Dense(3, activation = 'softmax'))
    # Compile model
    model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    return model

model = KerasClassifier(build_fn = create_model, epochs = 25, verbose = 0) 
                                                                      
# Building a simple search grid that adjusts nodes in hidden layers
param_grid = dict(hiddennodes = [500, 1000])

grid = GridSearchCV(estimator = model, param_grid = param_grid, cv = 5)
grid_result = grid.fit(X_train, y_train_labels, batch_size = 500)

# grid_result.cv_results_ for full results file
print('Best: %f using %s' % (grid_result.best_score_, grid_result.best_params_))
```

```python
# Compute test set accuracy
# Make predictions using best model
# y_preds = grid_result.best_estimator_.model.predict(X_test)
# best_y_preds_classindex = np.argmax(grid_result.best_estimator_.model.predict(X_test), axis = -1)

# Compute test set accuracy  
best_model_loss, best_model_acc = grid_result.best_estimator_.model.evaluate(X_test, y_test_labels)

print('Accuracy of best model on test set:', best_model_acc)
```

```python
# Get the class of prediction (by index)
y_preds_classindex_2 = np.argmax(grid_result.best_estimator_.model.predict(X_test), axis = -1)

# Transform the y_test labels from their one-hot encoded version
y_test_labels_2 = np.argmax(y_test_labels, axis = 1)

print('Multilayer Perceptron Neural Network')
from sklearn.metrics import classification_report

print(classification_report(y_test_labels_2, y_preds_classindex_2))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

#print(rf_model_opt.score(X_train_res, y_train_res))
print(accuracy_score(y_test_labels_2, y_preds_classindex_2))
print(f1_score(y_test_labels_2, y_preds_classindex_2, average='weighted'))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

#from keras import models
#from keras import layers

from sklearn.model_selection import train_test_split
#from tensorflow.keras.utils import to_categorical
import tensorflow as tf
#
# fix random seed for reproducibility
seed = 123
np.random.seed(seed)
tf.random.set_seed(seed)

model = Sequential([
    Dense(500, input_shape=(307,)), # input layer is embedded in first hidden layer 
    Activation('relu'), 
    Dense(250), # add another hidden layer
    Activation('relu'), 
    Dense(3), 
    Activation('softmax'),
])

model.summary()

# Once your model looks good, configure its learning process with .compile():
model.compile(loss = 'categorical_crossentropy', optimizer = 'sgd', metrics = ['accuracy'])

# Create training and test split
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y, random_state=42)
#
# Create categorical labels
#
y_train_labels = pd.get_dummies(y_train).values
y_test_labels = pd.get_dummies(y_test).values
#
# Fit the neural network
#
model.fit(X_train, y_train_labels, epochs = 50, batch_size = 50, validation_split = 0.2)
```

```python
# Get the accuracy of test data set
#
test_loss, test_acc = model.evaluate(X_test, y_test_labels)
#
# Print the test accuracy
print('Test Accuracy: ', test_acc, '\nTest Loss: ', test_loss)
```
